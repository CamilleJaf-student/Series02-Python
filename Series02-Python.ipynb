### Add a Markdown cell with the maths to generate a sine wave

 y(t)= A ⋅ sin(ωt+φ)

 y(t) : displacement, unit meter [m]

A : amplitude, unit meter [m]

ω : angular frequency, unit radian per second [rad·s⁻¹]

t : time, unit second [s]

φ : phase shift, unit radian [rad]
Whole equation unit: meter [m]

```python
# import libraries for generate sine function
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Create numpy array named t with 1000 values from 0 to 1
t = np.linspace(0, 1, 1000)

# Create numpy array named s that contains values of a sine function of t (s is for signal)
s = np.sin(2 * np.pi * t) 

# Plot s as a function of t and save it as a png file
plt.plot(t, s)
plt.title('Sine function')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
plt.savefig('sine_function_with_noise.png')

```
